## What is an Awkward array?

https://awkward-array.org/doc/main/getting-started/what-is-an-awkward-array.html

```python
ak.Array([1, 2, 3])
```

```python
ak.Array([
    [1, 2, 3],
    [4, 5, 6]
])
```

```python
ak.Array([
    [1, 2, 3],
    [4],
    [5, 6]
])
```

```python
ak.Array([
    [1, 2, 3],
    [4],
    [5, 6, None]
])
```

```python
ak.Array([
    [3, 141], 
    [59, 26, 535], 
    [8]
])
```
